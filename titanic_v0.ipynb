**Survival by Gender + Encoding**

```python
def sex_encoder(data):
  mapping = {"male": 1, "female": 0}
  data["Sex"] = data["Sex"].replace(mapping)
  return data

sex_encoder(train)
```

```python
sex_encoder(test)
```

```python
cp = sns.countplot(x=train["Sex"], hue=train["Survived"], data=train)
cp.set(xlabel="0 = female | 1 = male")
```

**Survival by Port Embarked + Encoding**

```python

print("Missing vales:",train["Embarked"].isna().sum())
sns.countplot(x=train["Embarked"], hue=train["Survived"], data=train)

#since S is the most frequent, we impute the mising value in test with S
train["Embarked"].fillna(value="S", inplace=True)
print("Missing vales after imputing:",train["Embarked"].isna().sum())

#encoding categorical data
def embarked_encoder(data):
  df = pd.get_dummies(data=data["Embarked"], prefix="Embarked")
  data = pd.concat([data, df], axis=1)
  data.drop(["Embarked"], axis=1, inplace=True)
  return data

train=embarked_encoder(train)
test=embarked_encoder(test)
train.head()
```

```python
test.head()
```

**Survival by Age + Impute missing values**

```python
train["Age"].describe()
```

```python
#UNCOMMENT WHEN RUNNING AGAIN
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
#replace nan with mean
train["Age"]=imputer.fit_transform(train[["Age"]])
test["Age"]=imputer.fit_transform(test[["Age"]])
#convert data type from series to int
train["Age"] = train["Age"].astype(int)
test["Age"] = test["Age"].astype(int)
#visualize age ranges
# hist = train["Age"].hist()
sns.histplot(data = train, x=train["Age"], kde=True, hue=train["Survived"])
#verify that there are no more missing values
print("Missing values after imputing: ",train["Age"].isnull().sum())
```

# **Model Training**

**Random Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(bootstrap=True,class_weight='balanced_subsample',criterion='gini',max_features='log2',n_estimators=100).fit(train,target)
Ypred=model.predict(test)
model.predict_proba(test)
```

**Model Evaluation Metrics**

*Cross Validation*

```python
from sklearn.model_selection import KFold,cross_val_score
kf=KFold(n_splits=25,shuffle=True)
score=cross_val_score(model,train,target,cv=kf,n_jobs=-1,verbose=0,fit_params=None)
score.mean()
```
